## Problema 1

```python
img = cv2.imread('paris.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
def canales_color_imagen(imagen, color):
  alto = img.shape[0]
  ancho = img.shape[1]
  imgSalida = np.zeros((alto, ancho, 3))
  for i in range(0, alto):
    for j in range(0, ancho):
      pixel = img[i, j]
      blue = pixel[2]
      green = pixel[1]
      red = pixel[0]
      if color == 1:    # Activo solo canal azul
        imgSalida[i, j] = [blue, 0, 0]
      elif color == 2:  # Activo solo canal verde
        imgSalida[i, j] = [0, green, 0]
      elif color == 3:  # Activo solo canal rojo
        imgSalida[i, j] = [0, 0, red]
      elif color == 10: # Activo solo canal rojo y verde
        imgSalida[i, j] = [0, green, red]
      elif color == 20: # Activo solo canal verde y azul
        imgSalida[i, j] = [blue, green, 0]
      elif color == 30: # Activo solo canal azul y rojo
        imgSalida[i, j] = [blue, 0, red]
      else:
        imgSalida[i, j] = [blue, green, red]
  return imgSalida
```

```python
def unir_canales_color(img_gray_red, img_gray_green, img_gray_blue):
  alto = img_gray_blue.shape[0]
  ancho = img_gray_blue.shape[1]
  imgSalida = np.zeros((alto, ancho, 3))
  for i in range(0, alto):
    for j in range(0, ancho):
      pixel_red = img_gray_red[i,j]
      pixel_green = img_gray_green[i,j]
      pixel_blue = img_gray_blue[i,j]
      imgSalida[i, j] = [pixel_blue[2], pixel_green[1], pixel_red[0]]
  return imgSalida
```

```python
def convertir_escala_grises(img, inc_gray_aritm = False):
  alto = img.shape[0]
  ancho = img.shape[1]
  gray_green = np.zeros((alto, ancho, 1))
  gray_blue = np.zeros((alto, ancho, 1))
  gray_red = np.zeros((alto, ancho, 1))
  gray_aritm = np.zeros((alto, ancho, 1))

  for i in range(0, alto):
    for j in range(0, ancho):
      pixel = img[i,j]
      gray_blue[i,j] = pixel[2]
      gray_green[i,j] = pixel[1]
      gray_red[i,j] = pixel[0]
      gray_aritm[i,j] = np.mean(pixel)
  if inc_gray_aritm:
    return gray_green, gray_blue, gray_red, gray_aritm
  else:
    return gray_green, gray_blue, gray_red
```

```python
img = cv2.imread('paris.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray_green, img_gray_blue, img_gray_red = convertir_escala_grises(img)
cv2.imwrite("Imagen_p3_salida_gray_verde.jpg", img_gray_green)
cv2.imwrite("Imagen_p3_salida_gray_azul.jpg", img_gray_blue)
cv2.imwrite("Imagen_p3_salida_gray_rojo.jpg", img_gray_red)
```

```python
img_gray_red = cv2.imread('Imagen_p3_salida_gray_rojo.jpg')
img_gray_red = cv2.cvtColor(img_gray_red, cv2.COLOR_BGR2RGB)
img_gray_green = cv2.imread('Imagen_p3_salida_gray_verde.jpg')
img_gray_green = cv2.cvtColor(img_gray_green, cv2.COLOR_BGR2RGB)
img_gray_blue = cv2.imread('Imagen_p3_salida_gray_azul.jpg')
img_gray_blue = cv2.cvtColor(img_gray_blue, cv2.COLOR_BGR2RGB)
imgSalida_3 = unir_canales_color(img_gray_red, img_gray_green, img_gray_blue)
cv2.imwrite('Imagen_Salida3.jpg', imgSalida_3)
imgTemp3 = cv2.imread('Imagen_Salida3.jpg')
imgTemp3 = cv2.cvtColor(imgTemp3, cv2.COLOR_BGR2RGB)
plt.imshow(imgTemp3)
```

## Problema 4

```python
def plot_histogramas(img, together=True):
  alto = img.shape[0]
  ancho = img.shape[1]

  img_gray_green, img_gray_blue, img_gray_red, img_gray_aritm = convertir_escala_grises(img, True)
  
  y_green = np.zeros((256), np.uint64)
  y_blue = np.zeros((256), np.uint64)
  y_red = np.zeros((256), np.uint64)
  y_aritm = np.zeros((256), np.uint64)

  for i in range(0,alto):
    for j in range(0,ancho):
        y_green[int(img_gray_green[i,j])] += 1
        y_blue[int(img_gray_blue[i,j])] += 1
        y_red[int(img_gray_red[i,j])] += 1
        y_aritm[int(img_gray_aritm[i,j])] += 1
  x = np.arange(0,256)
  mean_green = np.matmul(x+1, y_green)/sum(y_green)
  mean_blue = np.matmul(x+1, y_blue)/sum(y_blue)
  mean_red = np.matmul(x+1, y_red)/sum(y_red)
  mean_aritm = np.matmul(x+1, y_aritm)/sum(y_aritm)

  if together:
    plt.title('Histogramas de la imagen')
    plt.axvline(x=mean_green, color="green")
    plt.bar(x,y_green,color="green",align="center")
    plt.axvline(x=mean_blue, color="blue")
    plt.bar(x,y_blue,color="blue",align="center")
    plt.axvline(x=mean_red, color="red")
    plt.bar(x,y_red,color="red",align="center")
    plt.axvline(x=mean_aritm, color="gray")
    plt.bar(x,y_aritm,color="gray",align="center")
    plt.show()
  else:
    plt.title('Histograma color verde')
    plt.axvline(x=mean_green, color="black")
    plt.bar(x,y_green,color="green",align="center")
    plt.show()
    plt.title('Histograma color azul')
    plt.axvline(x=mean_blue, color="black")
    plt.bar(x,y_blue,color="blue",align="center")
    plt.show()
    plt.title('Histograma color rojo')
    plt.axvline(x=mean_red, color="black")
    plt.bar(x,y_red,color="red",align="center")
    plt.show()
    plt.title('Histograma escala de grises promedio aritmetico')
    plt.axvline(x=mean_aritm, color="black")
    plt.bar(x,y_aritm,color="gray",align="center")
    plt.show()
```

```python
plot_histogramas(img, False)
```

## Problema 5

```python
def escala_gris_ponderada(img, w_red=0.299, w_green=0.587, w_blue=0.11):
  alto = img.shape[0]
  ancho = img.shape[1]
  gray_ponderado = np.zeros((alto, ancho, 1))
  for i in range(0, alto):
    for j in range(0, ancho):
      pixel = img[i,j]
      gray_ponderado[i,j] = int(w_red*pixel[2] + w_green*pixel[1] + w_blue*pixel[0])
  return gray_ponderado
```

```python
img = cv2.imread('paris.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray_ponderado = escala_gris_ponderada(img)
cv2.imwrite("Imagen_p5_salida_gray_ponderado.jpg", img_gray_ponderado)
imgTemp5 = cv2.imread('Imagen_p5_salida_gray_ponderado.jpg')
imgTemp5 = cv2.cvtColor(imgTemp5, cv2.COLOR_BGR2RGB)
plt.imshow(imgTemp5)
```

### Espacio de color HSV

Los dispositivos de captura y visualización de imágenes captan y generan, respectivamente, el color mezclando los denominados colores primarios (rojo, verde y azul) con diferentes intensidades. Las mezclas resultantes constituyen una amplia variedad de colores, sin embargo, la relación entre la cantidad de rojo, verde y azul, y el color resultante no es intuitiva para el ojo humano. Ni los colores primarios (también conocidos como aditivos) ni los secundarios (también conocidos como sustractivos) definen las relaciones entre colores como el ojo humano, porque éste no percibe de la misma forma las intensidades de los colores rojo, verde y azul.

Para corregir este ‘defecto’ se crearon otros espacios de color como el HSV. Por sus siglas en inglés, HSV significa Hue, Saturation, Value o Tono Saturación, Valor, en español. 

Este modelo se basa en cómo se organizan los colores al ser percibidos por los humanos en términos de los atributos que definen el color (tono, croma y luminosidad), y también en los métodos tradicionales de mezcla de pigmento, esto consistente en mezclar pigmentos de color con pigmentos negros o blancos para conseguir pigmentos más oscuros o claros respectivamente.

El espacio de color HSV es la representación cilíndrica del cubo de color RGB (representación cartesiana).

image.png

El modelo HSV se denomina a veces HSB o Hue, Saturation, Brightness - Tono, Saturación, Brillo. En todos estos espacios, el tono corresponde al ángulo alrededor del eje del cilindro, la saturación es la distancia medida perpendicularmente al eje y el valor o intensidad se mide a lo largo del eje, como se muestra en la imagen anterior.

En estos espacios, los colores neutrales (grises o acromáticos) se localizan sobre el eje central vertical. Los colores primarios y secundarios se disponen alrededor del contorno exterior del cilindro, es decir, donde la saturación es igual a 1. En el espacio HSV los colores primarios y secundarios tienen valor igual a 1 (V = 1). Para este modelo, si se mezclan los colores *puros* con blanco, se reduce la saturación, mientras que si se mezclan con negro, la saturación se mantiene (S = 1).

Como estas definiciones de la saturación en las cuales los colores casi neutrales muy oscuros se consideran totalmente saturados, entran en conflicto con la noción intuitiva de la pureza del color, por lo que, se suelen representar estos espacios como un cono.
